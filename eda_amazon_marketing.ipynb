# EDA: Amazon Ecommerce Marketing Data (January 2020)

Quick exploratory data analysis of the Amazon marketing sample dataset covering January 1-31, 2020.

## 1. Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Set visualization style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
```

## 2. Load and Inspect the Dataset

```python
# Load the dataset
data_path = Path('/Users/blag/Documents/UChicago MS/2025 Fall/GenAI_finalProject/voice_to_voice_agentic_rag/data/marketing_sample_for_amazon_com-ecommerce__20200101_20200131__10k_data.csv')
df = pd.read_csv(data_path)

print("Dataset shape:", df.shape)
print("\nFirst 5 rows:")
df.head()
```

## 3. Data Shape and Info

```python
print("Data Types and Info:")
print(df.info())
print("\n" + "="*80)
print("\nColumn Names:")
print(df.columns.tolist())
```

## 4. Statistical Summary

```python
print("Descriptive Statistics:")
df.describe()
```

## 5. Missing Values Analysis

```python
# Check for missing values
missing_data = df.isnull().sum()
missing_percent = (missing_data / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing_Count': missing_data,
    'Percentage': missing_percent
}).sort_values('Missing_Count', ascending=False)

print("Missing Values:")
print(missing_df[missing_df['Missing_Count'] > 0])

if missing_df['Missing_Count'].sum() == 0:
    print("No missing values found!")

# Visualize missing values
fig, ax = plt.subplots(figsize=(12, 6))
missing_df[missing_df['Missing_Count'] > 0]['Percentage'].plot(kind='barh', ax=ax, color='coral')
ax.set_xlabel('Percentage of Missing Data (%)')
ax.set_title('Missing Values by Column')
plt.tight_layout()
plt.show()
```

## 6. Categorical Variables Distribution

```python
# Identify categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns

print(f"Categorical columns: {list(categorical_cols)}\n")

# Display unique values count for each categorical column
for col in categorical_cols:
    print(f"{col}: {df[col].nunique()} unique values")
    print(df[col].value_counts().head(10))
    print("\n" + "="*80 + "\n")
```

```python
# Create bar plots for categorical variables with limited unique values
categorical_cols_to_plot = [col for col in categorical_cols if df[col].nunique() <= 15]

fig, axes = plt.subplots(len(categorical_cols_to_plot), 1, figsize=(12, 3*len(categorical_cols_to_plot)))
if len(categorical_cols_to_plot) == 1:
    axes = [axes]

for idx, col in enumerate(categorical_cols_to_plot):
    df[col].value_counts().plot(kind='bar', ax=axes[idx], color='skyblue')
    axes[idx].set_title(f'Distribution of {col}')
    axes[idx].set_ylabel('Count')
    axes[idx].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

## 6.5 Product Category Distribution Analysis

```python
# Analyze product category distribution
print("="*80)
print("PRODUCT CATEGORY DISTRIBUTION ANALYSIS")
print("="*80)

# Basic statistics
print(f"\nTotal unique categories: {df['Category'].nunique()}")
print(f"Non-null categories: {df['Category'].notna().sum()}")
print(f"Missing categories: {df['Category'].isna().sum()}")

# Get top categories
category_counts = df['Category'].value_counts()
print(f"\nTop 15 Product Categories:")
print(category_counts.head(15))

# Calculate percentage distribution for top categories
print(f"\nTop 10 Categories (with percentages):")
top_10_cats = category_counts.head(10)
for cat, count in top_10_cats.items():
    percentage = (count / len(df)) * 100
    print(f"  {cat}: {count} products ({percentage:.2f}%)")
```

```python
# 1. Top 15 categories - Horizontal bar chart
plt.figure(figsize=(12, 8))
top_15 = df['Category'].value_counts().head(15)
plt.barh(range(len(top_15)), top_15.values, color='steelblue')
plt.yticks(range(len(top_15)), top_15.index, fontsize=10)
plt.xlabel('Number of Products', fontsize=11)
plt.title('Top 15 Product Categories by Count', fontsize=13, fontweight='bold')
plt.gca().invert_yaxis()
# Add value labels
for i, v in enumerate(top_15.values):
    plt.text(v + 10, i, str(v), va='center', fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# 2. Top 10 categories - Pie chart
plt.figure(figsize=(12, 8))
top_10 = df['Category'].value_counts().head(10)
other_count = df['Category'].value_counts()[10:].sum()
pie_data = list(top_10.values) + [other_count]
pie_labels = list(top_10.index) + ['Others']
colors = plt.cm.Set3(range(len(pie_data)))
plt.pie(pie_data, labels=pie_labels, autopct='%1.1f%%', colors=colors, startangle=90, textprops={'fontsize': 10})
plt.title('Top 10 Categories Distribution (% of Total)', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# 3. Distribution shape - Categories with at least 50 products
plt.figure(figsize=(12, 6))
categories_min_50 = df['Category'].value_counts()[df['Category'].value_counts() >= 50]
plt.bar(range(len(categories_min_50)), categories_min_50.values, color='coral', edgecolor='black', alpha=0.7)
plt.xlabel('Category Index', fontsize=11)
plt.ylabel('Number of Products', fontsize=11)
plt.title(f'Categories with â‰¥50 Products (Total: {len(categories_min_50)} categories)', fontsize=13, fontweight='bold')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# 4. Cumulative distribution
plt.figure(figsize=(12, 6))
sorted_counts = df['Category'].value_counts().sort_values(ascending=False)
cumsum = sorted_counts.cumsum()
cumsum_pct = (cumsum / len(df)) * 100
plt.plot(range(len(cumsum_pct)), cumsum_pct.values, marker='o', linewidth=2, markersize=4, color='darkgreen')
plt.axhline(y=80, color='r', linestyle='--', label='80% threshold')
plt.axhline(y=95, color='orange', linestyle='--', label='95% threshold')
plt.xlabel('Number of Categories', fontsize=11)
plt.ylabel('Cumulative % of Products', fontsize=11)
plt.title('Cumulative Distribution of Categories', fontsize=13, fontweight='bold')
plt.legend(fontsize=10)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Calculate insights
print(f"\nðŸ“Š KEY INSIGHTS:")
print(f"  - Total number of categories: {df['Category'].nunique()}")
print(f"  - Categories with â‰¥50 products: {len(categories_min_50)}")
print(f"  - Top 5 categories represent {(top_10.head(5).sum() / len(df) * 100):.1f}% of all products")
print(f"  - Top 10 categories represent {(top_10.sum() / len(df) * 100):.1f}% of all products")

# Find how many categories needed to reach 80% and 95% of products
cumsum_pct = (sorted_counts.cumsum() / len(df)) * 100
cats_for_80 = (cumsum_pct <= 80).sum()
cats_for_95 = (cumsum_pct <= 95).sum()
print(f"  - {cats_for_80} categories needed to represent 80% of products")
print(f"  - {cats_for_95} categories needed to represent 95% of products")
```

## 7. Numerical Variables Distribution

```python
# Identify numerical columns
numerical_cols = df.select_dtypes(include=[np.number]).columns

print(f"Numerical columns: {list(numerical_cols)}\n")

# Create histograms and box plots for numerical variables
fig, axes = plt.subplots(len(numerical_cols), 2, figsize=(14, 4*len(numerical_cols)))
if len(numerical_cols) == 1:
    axes = axes.reshape(1, -1)

for idx, col in enumerate(numerical_cols):
    # Histogram
    axes[idx, 0].hist(df[col].dropna(), bins=30, color='skyblue', edgecolor='black')
    axes[idx, 0].set_title(f'Distribution of {col}')
    axes[idx, 0].set_xlabel(col)
    axes[idx, 0].set_ylabel('Frequency')
    
    # Box plot
    axes[idx, 1].boxplot(df[col].dropna())
    axes[idx, 1].set_title(f'Box Plot of {col}')
    axes[idx, 1].set_ylabel(col)

plt.tight_layout()
plt.show()
```

## 8. Correlation Analysis

```python
# Compute correlation matrix for numerical columns
if len(numerical_cols) > 1:
    correlation_matrix = df[numerical_cols].corr()
    
    # Display correlation matrix
    print("Correlation Matrix:")
    print(correlation_matrix)
    
    # Visualize correlation matrix
    fig, ax = plt.subplots(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
                center=0, square=True, ax=ax, cbar_kws={'label': 'Correlation'})
    ax.set_title('Correlation Matrix of Numerical Variables')
    plt.tight_layout()
    plt.show()
else:
    print("Only one numerical column found, correlation analysis skipped.")
```

## 9. Time Period Analysis

```python
# Look for date columns and analyze temporal patterns
print("Looking for date-related columns...\n")

# Check if any columns might contain dates
date_cols = []
for col in df.columns:
    if 'date' in col.lower() or 'time' in col.lower():
        date_cols.append(col)
        print(f"Found potential date column: {col}")
        print(f"Sample values: {df[col].head()}")
        print()

# If no explicit date columns, check first column or describe the data range
if not date_cols:
    print("No explicit date columns found.")
    print("Dataset covers: January 1-31, 2020")
    print(f"Total rows: {len(df)}")
    print(f"Sample of all columns and their first row:")
    print(df.iloc[0])
```
